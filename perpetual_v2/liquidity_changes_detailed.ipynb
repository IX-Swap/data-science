```python
liquidity_changes_df = pd.read_csv("liquidity_changes.csv")
liquidity_changes_df.drop(columns=["Unnamed: 0"], inplace=True)
liquidity_changes_df["pool_name"] = liquidity_changes_df["base_token"].replace(tokens_addresses_to_names_dict)
liquidity_changes_df["time"] = pd.to_datetime(liquidity_changes_df["timestamp"], unit='s')
liquidity_changes_df
```

```python
fig = plt.figure(figsize=(25, 25))
index = 1

for pool in liquidity_changes_df["pool_name"].unique():
    token_liquidity_changes_df = liquidity_changes_df[liquidity_changes_df["pool_name"] == pool]
    ax = fig.add_subplot(4, 5, index)
    ax.plot(token_liquidity_changes_df["time"], token_liquidity_changes_df["upper_tick"], color='r', label='upper tick')
    ax.plot(token_liquidity_changes_df["time"], token_liquidity_changes_df["lower_tick"], color='g', label='lower tick')
    ax.set_title("Ticks of the " + pool + " pool")
    ax.legend()
    plt.xticks(rotation=45)
    index += 1
    
fig.tight_layout()
plt.show()
```
