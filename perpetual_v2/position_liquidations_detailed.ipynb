```python
position_liquidations_df = pd.read_csv("position_liquidations.csv")
position_liquidations_df.drop(columns=["Unnamed: 0"], inplace=True)
position_liquidations_df["pool_name"] = position_liquidations_df["base_token"].replace(tokens_addresses_to_names_dict)
position_liquidations_df["time"] = pd.to_datetime(position_liquidations_df["timestamp"], unit='s')
position_liquidations_df
```

```python
fig = plt.figure(figsize=(25, 25))
index = 1

for pool in position_liquidations_df["pool_name"].unique():
    pool_position_liquidations_df = position_liquidations_df[position_liquidations_df["pool_name"] == pool]
    ax = fig.add_subplot(4, 5, index)
    ax.plot(pool_position_liquidations_df["time"], pool_position_liquidations_df["liquidation_fee"], color='b')
    ax.set_xlabel("time")
    ax.set_ylabel("fee in USD")
    ax.set_title("Liquidation fee over time for " + pool + " pool")
    plt.xticks(rotation=45)
    index += 1
    
fig.tight_layout()
plt.show()
```

```python
position_liquidations_df["day_of_year"] = position_liquidations_df["time"].dt.day_of_year
```

```python
fig = plt.figure(figsize=(25, 25))
index = 1

for pool in position_liquidations_df["pool_name"].unique():
    pool_position_liquidations_df = position_liquidations_df[position_liquidations_df["pool_name"] == pool]
    daily_pool_position_liquidations_df = pool_position_liquidations_df[["day_of_year", "liquidation_fee"]
                                                                        ].groupby("day_of_year").sum()
    ax = fig.add_subplot(4, 5, index)
    ax.plot(daily_pool_position_liquidations_df[daily_pool_position_liquidations_df.index < 200].index, 
            daily_pool_position_liquidations_df[daily_pool_position_liquidations_df.index < 200]["liquidation_fee"], color='g')
    ax.set_xlabel("time")
    ax.set_ylabel("fee in USD")
    ax.set_title("Liquidation fee daily sum over time for " + pool + " pool")
    plt.xticks(rotation=45)
    index += 1
    
fig.tight_layout()
plt.show()
```
