```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
def show_value_price_timeseries(
    df: pd.DataFrame, 
    plot_name: str, price_instead_value: bool,
    x_size: int=15, y_size: int=9
):
    # set size and set timeseries representation with automatic definition
    fig, ax = plt.subplots(figsize=(x_size, y_size))
    fig.autofmt_xdate()

    if price_instead_value:
        ax.plot_date(df.datetime, df.Price, linestyle='solid', marker=None, color='red', label=str("Share price"))
    else:
        ax.plot_date(df.datetime, df.Value, linestyle='solid', marker=None, label=str("Trades value"))

    # set labels
    ax.set_xlabel("Time")
    ax.set_ylabel("Price and value")

    # set title and legend
    fig.suptitle(plot_name)
    fig.legend()
```
