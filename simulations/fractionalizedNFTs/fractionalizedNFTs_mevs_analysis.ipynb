## DOG/WETH MEV attacks analysis

```python
dog_price_df = dog_weth_sim.get_daily_token_price(first_to_second=False, token_name='DOG')
dog_price_df['date'] = dog_price_df['date'].dt.strftime("%Y-%m-%d")
dog_price_df
```

```python
dog_weth_mevs_df = pd.merge(dog_weth_mevs_df, dog_price_df, how='left', on='date')
```

```python
dog_weth_mevs_df['bruto_weth_to_usdc_profit'] = dog_weth_mevs_df['WETH_profit'] * dog_weth_mevs_df['WETH_price']
dog_weth_mevs_df['bruto_total_profit'] = dog_weth_mevs_df['bruto_weth_to_usdc_profit'] + (dog_weth_mevs_df['DOG_profit'] *
                                                                                          dog_weth_mevs_df['DOG_price'])
dog_weth_mevs_df.loc[dog_weth_mevs_df['tx_fee'] >= 1, 'tx_fee'] = dog_weth_mevs_df['tx_fee'].median()
dog_weth_mevs_df.loc[dog_weth_mevs_df['tx_fee'] == 0, 'tx_fee'] = dog_weth_mevs_df['tx_fee'].median()
dog_weth_mevs_df['gas_costs_usd'] = (dog_weth_mevs_df['tx_fee'] * dog_weth_mevs_df['WETH_price'] *
                                      dog_weth_mevs_df['gas_usage_by_tx'])
dog_weth_mevs_df['netto_profit'] = dog_weth_mevs_df['bruto_total_profit'] - dog_weth_mevs_df['gas_costs_usd']
dog_weth_mevs_df.dropna(inplace=True)
dog_weth_mevs_df
```

```python
dog_weth_count_df = dog_weth_mevs_df.value_counts(subset=['timestamp'])
dog_weth_count_df[dog_weth_count_df.values > 2]
```

```python
grouped_dog_weth_profits_df = dog_weth_mevs_df.groupby(by=['timestamp']).sum()
grouped_dog_weth_profits_df.drop(columns=['block_position', 'tx_fee'], inplace=True)
grouped_dog_weth_profits_df['block'] = grouped_dog_weth_profits_df['block'] / 2
grouped_dog_weth_profits_df['bruto_total_profit'] = grouped_dog_weth_profits_df['bruto_total_profit']/2
grouped_dog_weth_profits_df['netto_profit'] = (grouped_dog_weth_profits_df['bruto_total_profit'] -
                                               grouped_dog_weth_profits_df['gas_costs_usd'])
grouped_dog_weth_profits_df.reset_index(inplace=True)

print(fr"max brutto profit {grouped_dog_weth_profits_df['bruto_total_profit'].max()}")
print(fr"min brutto profit {grouped_dog_weth_profits_df['bruto_total_profit'].min()}")
print(fr"max netto profit {grouped_dog_weth_profits_df['netto_profit'].max()}")
print(fr"min netto profit {grouped_dog_weth_profits_df['netto_profit'].min()}")
print(fr"zero bruto profit transactions = {len(dog_weth_mevs_df[dog_weth_mevs_df['bruto_total_profit'] == 0])}")
print(fr"transactions with losses = {len(dog_weth_mevs_df[dog_weth_mevs_df['netto_profit'] < 0])}")
```

```python
print(fr"sum of paid gas = {sum(dog_weth_mevs_df['gas_costs_usd'])}")
print(fr"sum of brutto profits = {sum(dog_weth_mevs_df['bruto_total_profit'] / 2)}")
print(fr"sum of netto profits = {sum(dog_weth_mevs_df['bruto_total_profit'] / 2) - sum(dog_weth_mevs_df['gas_costs_usd'])}")
```

```python
dog_weth_swaps_df = dog_weth_sim.get_original_swaps_df()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (15, 10))
ax1.plot(grouped_dog_weth_profits_df['timestamp'], grouped_dog_weth_profits_df['bruto_total_profit'], label='bruto profit', color='r')
ax1.plot(grouped_dog_weth_profits_df['timestamp'], grouped_dog_weth_profits_df['netto_profit'], label='netto profit', color='g', linestyle='--')
ax1.set_xlabel('time')
ax1.set_ylabel('USD profit')
ax1.set_title('Profits extracted by MEV attacks from DOG/WETH')
ax1.legend()

ax2.plot(dog_weth_swaps_df['timestamp'], dog_weth_swaps_df['amount_usd'], label="Uniswap tx capitalization", color='m')
ax2.set_xlabel('time')
ax2.set_ylabel('Capitalization in USD')
ax2.set_title('Capitalization of transactions from DOG/WETH')
ax2.legend()

plt.tight_layout()
plt.show()
```
