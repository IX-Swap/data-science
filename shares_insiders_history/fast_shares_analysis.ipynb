# Comparing Pareto, Cauchy distributions with shares distribution

````python
from scipy.stats import halfcauchy

# generate half-cauchy (generates non-negative values) and pareto distributions 
cauchy = halfcauchy.rvs(size=1312, loc=0, scale=1700)
pareto = np.random.pareto(0.258, size=1700)

plt.figure(figsize=(10, 10))
plt.subplots_adjust(wspace=0.1, hspace=0.25)

plt.subplot(3, 1, 1)
sns.histplot(x=cauchy, color='green', binrange=[0, 300000], bins=100).set_title('Cauchy distribution')

plt.subplot(3, 1, 2)
sns.histplot(x=pareto, color='blue', binrange=[0, 300000], bins=100).set_title('Pareto distribution')

plt.subplot(3, 1, 3)
sns.histplot(data=apple_df, x='Shares', color='gray').set_title('Apple shares transaction amount and their frequency')

plt.show()
````

# Advanced analysis (trying to improve linecharts for better overviews)

````python
def show_transactions_distribution_moving_average(
    df: pd.DataFrame, time_cycle: str, plot_name: str,
    target_column: str, x_size: int=15, y_size: int=9
    ) -> pd.DataFrame:
    
    # set new timestamp for making average calculations without touching original dataset
    # form column with second resolution
    df_time = df.copy()
    df_time.datetime = pd.to_datetime(df_time.datetime, unit='s')
    df_time = df_time[df_time[target_column] > 0]

    # set timestamp as index and then resample conform both resolutions (smaller_average 
    # step and bigger_average step)
    df_time.index = df_time.datetime
    df_avg_time = df_time.rolling(time_cycle).mean()
    df_time = df_avg_time

    # set size and set timeseries representation with automatic definition
    fig, ax = plt.subplots(figsize=(x_size, y_size))
    fig.autofmt_xdate()

    # plot information
    ax.plot_date(df_time.index, df_time[target_column], linestyle='solid', marker=None)

    # set labels
    ax.set_xlabel("Time")
    ax.set_ylabel("Average transaction value per " + time_cycle)

    # set title and legend
    fig.suptitle(plot_name)
    fig.legend()

    return df_time
````

## Google

Share price evolution shows clear rise of price over time. Considering that data contains drops related to gaps in insider trades, rise almost stable. Conform shares movement and capitalization graphs can be clearly seen that there was some strange rise in shares activity from the end of 2020 till October 2021 (before this moment there was a flat line of silence).

Count distribution by capitalization value shows that distribution by overall shape looks like exponential one (Pareto or Cauchy), but deviations from exponential shape are strong, reducing similarity.

````python
smooth_google_df = show_transactions_distribution_moving_average(
    google_df, time_cycle='30d', plot_name="Smoothed share price over time", target_column='Price',
    x_size=5, y_size=5
    )
````

````python
plt.figure(figsize=(5, 5))
ax = sns.lineplot(data=smooth_google_df, x='datetime', y='Value', color='red').set_title(
    'smoothed shares trades capitalization over time'
    )
plt.xticks(rotation=50)
plt.show()
````

````python
plt.figure(figsize=(5, 5))
sns.lineplot(data=smooth_google_df, x='datetime', y='Shares', color='green').set_title(
    'smoothed shares movements over time'
    )
plt.xticks(rotation=50)
plt.show()
````

````python
plt.figure(figsize=(5, 5))
sns.histplot(data=smooth_google_df, x='Value', color='orange', bins=50).set_title(
    'smoothed shares capitalizations count by value'
    )
plt.show()
````

````python
plt.figure(figsize=(5, 5))
sns.histplot(data=smooth_google_df, x='Shares', color='magenta', bins=50).set_title(
    'smoothed shares capitalizations count by value'
    )
plt.show()
````
