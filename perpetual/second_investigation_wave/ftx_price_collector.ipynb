## Get daily price data for specified market
### ETH price

```python
# resolution to get daily data
resolution = 60 * 60 * 24
start_datetime = datetime(2022, 1, 1).timestamp()
market_name = "ETH/USD"
```

```python
path = f'/markets/{market_name}/candles?resolution={resolution}&start={start_datetime}'
url = ENDPOINT + path
url
```

```python
res = requests.get(url).json()
df = pd.DataFrame(res['result'])
df['date'] = pd.to_datetime(df['startTime'])
df
```

```python
df.to_csv("ftx_price_for_eth.csv")
```

# Collecting reserves stories from the Uniswap

```python
weth_usdc_reserves_df = collect_uniswap_price_daily_df("0xb4e16d0168e52d35cacd2c6185b44281ec28c9dc") # WETH/USDC
wbtc_weth_reserves_df = collect_uniswap_price_daily_df("0xbb2b8038a1640196fbe3e38816f3e67cba72d940") # WBTC/WETH
perp_weth_reserves_df = collect_uniswap_price_daily_df("0xf66369997ae562bc9eec2ab9541581252f9ca383") # PERP/WETH
# there is no pool with DOT token on Uniswap
```

```python
weth_usdc_reserves_df["price"] = weth_usdc_reserves_df["reserve0"] / weth_usdc_reserves_df["reserve1"]
weth_usdc_reserves_df
```

```python
wbtc_weth_reserves_df["price_to_eth"] = wbtc_weth_reserves_df["reserve1"] / wbtc_weth_reserves_df["reserve0"]
wbtc_weth_reserves_df = pd.merge(wbtc_weth_reserves_df, weth_usdc_reserves_df[["date", "price"]], on=["date"])
wbtc_weth_reserves_df["price"] = wbtc_weth_reserves_df["price_to_eth"] * wbtc_weth_reserves_df["price"]
wbtc_weth_reserves_df
```

```python
perp_weth_reserves_df["price_to_eth"] = perp_weth_reserves_df["reserve1"] / perp_weth_reserves_df["reserve0"]
perp_weth_reserves_df = pd.merge(perp_weth_reserves_df, weth_usdc_reserves_df[["date", "price"]], on=["date"])
perp_weth_reserves_df["price"] = perp_weth_reserves_df["price_to_eth"] * perp_weth_reserves_df["price"]
perp_weth_reserves_df
```

```python
weth_usdc_reserves_df[["date", "price"]].to_csv("uniswap_price_for_weth.csv")
wbtc_weth_reserves_df[["date", "price"]].to_csv("uniswap_price_for_wbtc.csv")
perp_weth_reserves_df[["date", "price"]].to_csv("uniswap_price_for_perp.csv")
```
