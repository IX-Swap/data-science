# Collecting required parts

```python
def tx_version_separate(record):
    splitted_id = record['id'].split('-')
    if len(splitted_id) > 1:
        record['id'] = splitted_id[0]
        record['tx_change_version'] = splitted_id[1]
    
    return record
```

```python
%%time
master_story_funding_rate_part_df = funding_rate_events_df[['id', 'block_number', 'timestamp', 'amm']]
master_story_funding_rate_part_df['origin'] = 'funding rate'
master_story_funding_rate_part_df['tx_change_version'] = np.zeros(len(master_story_funding_rate_part_df))
master_story_funding_rate_part_df = master_story_funding_rate_part_df.apply(lambda record: tx_version_separate(record), axis=1)
master_story_funding_rate_part_df
```

```python
%%time
master_story_liquidations_part_df = liquidation_events_df[['id', 'block_number', 'timestamp', 'amm']]
master_story_liquidations_part_df['origin'] = 'liquidation'
master_story_liquidations_part_df['tx_change_version'] = np.zeros(len(master_story_liquidations_part_df))
master_story_liquidations_part_df = master_story_liquidations_part_df.apply(lambda record: tx_version_separate(record), axis=1)
master_story_liquidations_part_df
```

```python
%%time
master_story_margin_changed_part_df = margin_changed_events_df[['id', 'block_number', 'timestamp', 'amm']]
master_story_margin_changed_part_df['origin'] = 'margin change'
master_story_margin_changed_part_df['tx_change_version'] = np.zeros(len(master_story_margin_changed_part_df))
master_story_margin_changed_part_df = master_story_margin_changed_part_df.apply(lambda record: tx_version_separate(record), axis=1)
master_story_margin_changed_part_df
```

```python
%%time
master_story_position_changes_part_df = position_changed_events_df[['id', 'block_number', 'timestamp', 'amm']]
master_story_position_changes_part_df['origin'] = 'position change'
master_story_position_changes_part_df['tx_change_version'] = np.zeros(len(master_story_position_changes_part_df))
master_story_position_changes_part_df = master_story_position_changes_part_df.apply(lambda record: tx_version_separate(record), axis=1)
master_story_position_changes_part_df
```

```python
master_story_list = [master_story_funding_rate_part_df,
                    master_story_liquidations_part_df,
                    master_story_margin_changed_part_df,
                    master_story_position_changes_part_df]

master_story_df = pd.concat(master_story_list)
```

```python
master_story_df
```

```python
master_story_df.sort_values(by=['block_number', 'tx_change_version'], inplace=True)
master_story_df.reset_index(inplace=True)
```

```python
master_story_df.drop(columns=['index'], inplace=True)
```
