```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

```python
user_data_df = pd.read_csv('traders_detailed_info.csv')
user_data_df.drop(columns=['Unnamed: 0'], inplace=True)
user_data_df
```

```python
user_data_df.describe().T
```

```python
plt.figure(figsize=(12, 7))
sns.heatmap(user_data_df.corr(), annot=True)
plt.show()
```

```python
fig = plt.figure(figsize=(14, 14))

index = 1
for column in ["max_found_margin", "median_margin",
              "max_found_position_notional", "median_position_notional", "min_found_position_notional",
              "max_found_exchanged_position_size", "median_exchanged_position_size", "min_found_exchanged_position_size",
              "operations_count"]:
    ax = plt.subplot(3, 3, index)
    ax.hist(user_data_df[column], bins=100)
    ax.set_xlabel(column)
    ax.set_ylabel("Count of records")
    ax.set_title("Values distribution for " + column)
    ax.grid(True, linestyle='--')
    index += 1
    
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 14))

index = 1
for column in ["max_found_margin", "median_margin",
              "max_found_position_notional", "median_position_notional", "min_found_position_notional",
              "max_found_exchanged_position_size", "median_exchanged_position_size", "min_found_exchanged_position_size",
              "operations_count"]:
    ax = plt.subplot(3, 3, index)
    ax.hist(user_data_df[user_data_df['total_pnl'] > 0][column], bins=100)
    ax.set_xlabel(column)
    ax.set_ylabel("Count of records")
    ax.set_title("Values distribution for " + column)
    ax.grid(True, linestyle='--')
    index += 1
    
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(user_data_df["total_pnl"], bins=100)
ax.set_xlabel("profit and loss")
ax.set_ylabel("records count")
ax.set_title("distribution of profit and loss")
ax.grid(True, linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(user_data_df[user_data_df['total_pnl'] > 1000]["total_pnl"], bins=100)
ax.set_xlabel("profit and loss")
ax.set_ylabel("records count")
ax.set_title("distribution of profit and loss (only bigger than 1000 USD profit)")
ax.grid(True, linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(user_data_df[user_data_df['total_pnl'] < -1000]["total_pnl"], bins=100)
ax.set_xlabel("profit and loss")
ax.set_ylabel("records count")
ax.set_title("distribution of profit and loss (only losses bigger than 1000 USD)")
ax.grid(True, linestyle='--')
plt.show()
```

```python
print("there are " + str(len(user_data_df[user_data_df['total_pnl'] > 0])) + " traders who exctracted profit" + 
     " and " + str(len(user_data_df[user_data_df['total_pnl'] < 0])) + " who had losses and total collected " +
     "profit is " + str(user_data_df[user_data_df['total_pnl'] > 0]['total_pnl'].sum()) + ' while total losses ' +  
     'are equal to ' + str(user_data_df[user_data_df['total_pnl'] < 0]['total_pnl'].sum()) + ' and therefore' + 
     ' overall balance is equal to ' + str(user_data_df['total_pnl'].sum()))
```

```python
time_position_changes_df = pd.read_csv("opens_closes_with_time.csv")
time_position_changes_df
```
